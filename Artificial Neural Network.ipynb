```python
# Seperating Dependent and Independent Dataset 

dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.drop(['RowNumber','CustomerId','Surname','Exited'],axis=1)
y = dataset['Exited']
X
```

```python
# Dummy encoding for Categorical COlumns
X = pd.get_dummies(X,columns=['Geography','Gender'],drop_first=True)
```

```python
# Splitting Data into training Data and test set

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size= 0.2,random_state =0)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

### Building ANN

```python
# Initializing the ANN
ann = tf.keras.models.Sequential()
```

```python
# Bilding the input Layer and 1st Hidden Layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
# Bilding the 2nd Hidden Layer

ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
```

```python
# Building Outpt layer

ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
```

### Compiling The ANN

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics = ['accuracy'])
```

```python
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

### Predicting the Out put for Test set

```python
y_pred = ann.predict(X_test)
Y_pred_c = y_pred.round()
```

### Calculating the Accuracy of ANN

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_pred_c,y_test)
```
